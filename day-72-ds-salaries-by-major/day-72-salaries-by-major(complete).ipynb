```python
import requests
from bs4 import BeautifulSoup

header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9,he;q=0.8"
    }

response = requests.get(url='https://www.payscale.com/college-salary-report/majors-that-pay-you-back/bachelors', headers=header)
print(response.status_code)
with open('payscale_response.txt', 'w') as file:
    file.write(response.text)
```

```python
import pandas as pd
import re

with open('payscale_response.html', 'r') as file:
    text = file.read()

pattern = r'\D'

soup = BeautifulSoup(text, 'lxml')

for span in soup.find_all('span', class_='data-table__title'):
    span.decompose()


df = pd.read_html(str(soup))
df[0].rename(columns=lambda x: x.split('  ')[0], inplace=True)
# clean_df = df[0]['% High Meaning'].apply(lambda x: int(str(x).replace(',', '').replace('$', '')))
clean_df = df[0].copy()
columns_to_change = clean_df.columns[-3:]
for col in columns_to_change:
    clean_df[col] = clean_df[col].apply(lambda x: int(re.sub(pattern, '', str(x))) if bool(re.search(r'\d', str(x))) else 0)

spread_col = abs(clean_df[clean_df.columns[-3]] - clean_df[clean_df.columns[-2]])
clean_df.insert(6, 'Spread', spread_col)

print(clean_df)
```
